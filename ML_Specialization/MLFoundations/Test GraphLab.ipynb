```python
def tco(country):
    if country == 'USA':
        return 'United States'
    else:
        return country
```

```python
tco('Brazil')
```

```python
tco('USA')
```

```python
sf['Country'].apply(tco)
```

```python
sf['Country'] = sf['Country'].apply(tco)
```
