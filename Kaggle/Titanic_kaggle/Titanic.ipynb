```python
import graphlab
```

```python
sf = graphlab.SFrame('train.csv')
```

```python
sf.head()
```

```python
sf = sf.dropna()
train_data,val_data = sf.random_split(0.8,seed=20)
```

```python
graphlab.boosted_trees_classifier.get_default_options()
```

```python
model = graphlab.svm_classifier.create(sf,
                                                 target = 'Survived',
                                           max_iterations = 10000,
                                      penalty=1000000)
```

```python
print model.evaluate(sf)
```

```python
final_test = graphlab.SFrame('test.csv')
```

```python
pred = model.predict(final_test)
cp = final_test
cp.add_column(pred)
```
