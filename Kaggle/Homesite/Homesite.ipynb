```python
import graphlab
```

```python
sf = graphlab.SFrame('train.csv')
```

```python
sf.head()
```

```python
import random
x = random.randint(1,100)
train_data,test_data = sf.random_split(0.6,seed = x)
x = random.randint(1,100)
test_data,val_data = test_data.random_split(0.5,seed = x)
```

```python
model = graphlab.boosted_trees_classifier.create(train_data,
                                                 target='QuoteConversion_Flag',
                                                validation_set=val_data,
                                                max_iterations =1000)
```

```python
sf.remove_columns(['PropertyField6','GeographicField10A','GeographicField10B'])
```

```python
results = model.evaluate(test_data)
```
